```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Reset the index to make 'coin_id' a regular column
df_market_data = df_market_data.reset_index()

# Rename the 'index' column to 'coin_id' (if needed)
df_market_data.rename(columns={'index': 'coin_id'}, inplace=True)

# Display the updated DataFrame
df_market_data.head()
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---


### Prepare the Data


```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
columns_to_normalize = StandardScaler().fit_transform(df_market_data[[
    "price_change_percentage_24h",
    "price_change_percentage_7d",
    "price_change_percentage_14d",
    "price_change_percentage_30d",
    "price_change_percentage_60d",
    "price_change_percentage_200d",
    "price_change_percentage_1y"
]])
```

```python
# Create a DataFrame with the scaled data
df_scaled_data = pd.DataFrame(columns_to_normalize, columns=["price_change_percentage_24h",
    "price_change_percentage_7d",
    "price_change_percentage_14d",
    "price_change_percentage_30d",
    "price_change_percentage_60d",
    "price_change_percentage_200d",
    "price_change_percentage_1y"])


# Copy the crypto names from the original DataFrame
df_scaled_data["coin_id"] = df_market_data["coin_id"]

# Set the coin_id column as index
df_scaled_data.set_index("coin_id", inplace=True)

# Display the scaled DataFrame
df_scaled_data.head()
```

---


### Find the Best Value for k Using the Original Scaled DataFrame.


```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_scaled_data)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question:

**Question:** What is the best value for `k`?

**Answer:** In my opinion, 4 is the optimal value for k because it represents the point where the curve on the graph experiences the most significant bend in its trajectory.

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame


```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_scaled_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_model_predict = model.predict(df_scaled_data)

# Print the resulting array of cluster values.
k_model_predict
```

```python
# Create a copy of the scaled DataFrame
df_scaled_data_predict = df_scaled_data.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_data_predict["crypto_predictions"] = k_model_predict

# Display the copy of the scaled DataFrame
df_scaled_data_predict.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
# Select the features for clustering
features = df_scaled_data_predict[["price_change_percentage_24h", "price_change_percentage_7d"]]

# Fit K-Means with the desired number of clusters
kmeans = KMeans(n_clusters=4, random_state=42)
df_scaled_data_predict["cluster_labels"] = kmeans.fit_predict(features)
scatter_plot = df_scaled_data_predict.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="cluster_labels",          # Color points by cluster labels
    hover_cols=["coin_id"],      # Add crypto names as hover info
    title="Cryptocurrency Clustering",
    xlabel="24-Hour Price Change (%)",
    ylabel="7-Day Price Change (%)",
    cmap="Category10",           # Color map for clusters
    size=80,                     # Adjust point size
    height=500,
    width=700
)

# Display the scatter plot
scatter_plot
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.


```python
# Composite plot to contrast the Elbow curves
# YOUR CODE HERE!
import hvplot.pandas
from sklearn.cluster import KMeans
import pandas as pd

# Function to calculate inertia for a given dataset
def calculate_inertia(data, k_values):
    inertia_values = []
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        inertia_values.append(kmeans.inertia_)
    return inertia_values

# Define range of k values
k_values = range(1, 11)

# Calculate inertia for dataset1 (model_pca_df)
inertia_dataset1 = calculate_inertia(model_pca_df, k_values)

# Calculate inertia for dataset2 (df_scaled_data)
inertia_dataset2 = calculate_inertia(df_scaled_data, k_values)

# Create a DataFrame for plotting
elbow_df = pd.DataFrame({
    "k_values": list(k_values) * 2,
    "inertia": inertia_dataset1 + inertia_dataset2,
    "dataset": ["model_pca_df"] * len(k_values) + ["df_scaled_data"] * len(k_values)
})

# Plot the composite elbow curve using hvPlot
elbow_plot = elbow_df.hvplot.line(
    x="k_values",
    y="inertia",
    by="dataset",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    title="Elbow Curve Comparison",
    line_width=2,
    grid=True
)

elbow_plot
```

```python
# Composite plot to contrast the clusters
# YOUR CODE HERE!
from sklearn.cluster import KMeans
import pandas as pd

# Assuming 'df_scaled_data' and 'model_pca_df' are your dataframes

# Fit KMeans for the PCA model
kmeans_pca = KMeans(n_clusters=4, random_state=1)
kmeans_pca.fit(model_pca_df[['PCA1', 'PCA2', 'PCA3']])

# Fit KMeans for the scaled data
kmeans_scaled = KMeans(n_clusters=4, random_state=1)
kmeans_scaled.fit(df_scaled_data)

# Calculate inertia for both datasets (using inertia_)
inertia_pca = kmeans_pca.inertia_
inertia_scaled = kmeans_scaled.inertia_

# Create a new dataframe for plotting the elbow curve comparison
elbow_df = pd.DataFrame({
    "k_values": list(range(1, 2)) * 2,  # Since we are comparing two inertia values, just 1 value for each dataset
    "inertia": [inertia_pca] + [inertia_scaled],  # Add inertia values directly into a list
    "dataset": ["model_pca_df", "df_scaled_data"]  # Labels for datasets
})

# Plotting using hvplot to visualize the clusters and contrast them
import hvplot.pandas

# Composite plot for clusters
pca_plot = model_pca_df.hvplot.scatter(
    x='PCA1', 
    y='PCA2', 
    c=kmeans_pca.labels_, 
    colormap='viridis', 
    hover_cols=['coin_id'],
    title="Clusters from PCA Data",
    height=500,
    width=400
)

scaled_plot = df_scaled_data.hvplot.scatter(
    x='price_change_percentage_24h', 
    y='price_change_percentage_7d', 
    c=kmeans_scaled.labels_, 
    colormap='viridis', 
    hover_cols=['coin_id'],
    title="Clusters from Scaled Data",
    height=500,
    width=400
)

# Combine both plots into a composite plot
composite_plot = pca_plot + scaled_plot
composite_plot
```
